```python
## Merging DataFrame

df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Salary': [25, 30]})
df2 = pd.DataFrame({'Name': ['Alice', 'Mak'], 'Salary': [50000, 60000]})
```

```python
## Merge these dataframe on the "Name"

pd.merge(df1, df2, on='Name', how='inner')
```

```python
pd.merge(df1, df2, on='Name', how='outer')
```
