```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

im = cv2.imread("balmukund_pic.jpg")

print("orginal image")

# Show the result
#plt.figure(figsize=(12, 6))

#plt.subplot(1, 3, 1)
plt.title('Original Image')
#sns.plot(im)
plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))

# plt.plot(im)
plt.show()
print(im)

```

```python
blured_im = cv2.GaussianBlur(im, (3,3), sigmaX= 2)


plt.title(label="blurred image")
plt.imshow(cv2.cvtColor(blured_im, cv2.COLOR_BGR2RGB))
plt.show()

blured_im
```

```python
gray_scale_image = cv2.cvtColor(blured_im, cv2.COLOR_RGB2GRAY)

plt.title(label="grayscale image")
plt.imshow(gray_scale_image)
plt.show()
```
